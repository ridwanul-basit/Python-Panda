```python
import pandas as pd
```

# Way-1  Using Dictionary

```python
weather_data={
    "Day" : ["1/01/2025","10/01/2025","20/01/2025","31/01/2025","10/02/2025","20/02/2025"],
    "Temperature" : [10,12,11,15,13,20],
    "Windspeed" : [3,1,5,4,2,7],
    "Event" : ["Snow","Rain","Snow","Sunny","Sunny","Hot"]
}
df=pd.DataFrame(weather_data)
```

```python
df
```

# Way-2 Using Tuple in List

For this in a list here will be tuple where every tuple is a column and and after set columns we have to se columns in a different way which connected to creating DataFrame

```python
weather_data=[
    ("1/01/2025,10","Snow"),
     ("10/01/2025",12,"Rain"),
      ("20/01/2025",11,"Snow")
]
df=pd.DataFrame(weather_data, columns=["Day","Temperature","Event"])
df
```
